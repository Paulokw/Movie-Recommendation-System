```python
import pandas as pd
import difflib
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv(r"C:\Users\pc\Downloads\archive (15)\movies.csv")
df.head()
```

```python
features = ['genres','title','overview','tagline','cast','director']
```

```python
df[features]
```

```python
df2=df[features].fillna('')
```

```python
movies = df2['genres']+' '+df2['title']+' '+df2['overview']+' '+df2['tagline']+' '+df2['cast']+' '+df2['director']
movies
```

```python
movies.loc[4799]
```

```python
df.loc[4802]
```

```python
movies.tail()
```

```python
vec=TfidfVectorizer()
```

```python
feature_vectors=vec.fit_transform(movies)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity
```

```python
film=input('which movie would you like to watch: ')
```

```python
movie_list=df['title'].tolist()
movie_list
```

```python
close_match=difflib.get_close_matches(film,movie_list)
match=close_match[0]
```

```python
movies_index=df[df.title==match]['index'].values[0]
print(movies_index)
```

```python
similarity_score=list(enumerate(similarity[movies_index].tolist()))
print(similarity_score)
```

```python
sorted_sim_score=sorted(similarity_score, key=lambda x:x[1],reverse=True)
sorted_sim_score
```

```python
print('movies suggested for you: \n')
i=1
for sim in sorted_sim_score:
    index=sim[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if (i<10):
        print(i,'',title_from_index)
        i+=1
```

MOVIE RECOMMENDATION SYSTEM

```python
film=input('which movie would you like to watch: ')

movie_list=df['title'].tolist()

close_match=difflib.get_close_matches(film,movie_list)
match=close_match[0]

movies_index=df[df.title==match]['index'].values[0]

similarity_score=list(enumerate(similarity[movies_index].tolist()))

sorted_sim_score=sorted(similarity_score, key=lambda x:x[1],reverse=True)

print('movies suggested for you: \n')
i=1
for sim in sorted_sim_score:
    index=sim[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if (i<10):
        print(i,'',title_from_index)
        i+=1
```
